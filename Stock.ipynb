## BOND RETURN & VOLATILITY

```python
for stock in set(df['STOCK CODE']):
    
    df_cur = df.loc[df['STOCK CODE'] == stock, :]
    print(df_cur['NEXT COUPON RATE'])

    for index in range(df_cur.shape[0]):
        print(df_cur.iloc[index, df_cur.columns.get_loc('NEXT COUPON RATE')])

    
def get_yield_income(ncp, emp):
    return ncp / emp

def get_roll_down_return(empc, emp):
    return empc / emp

def get_expected_change_in_yield(md, emyc, c):
    return ((-(md) * emyc) + (0.5 * c * emyc * emyc))

def get_bond_return_of_month(yi, rdr, eciy):
    return yi + rdr + eciy
```

```python
import math

monthly_avg_return = []
volatility = []
stock_code = []
bond_over_volatility_ratio = []

for stock in set(df['STOCK CODE']):

    stock_code.append(stock)
    
    yield_income = []
    roll_down_return = []
    expected_change_in_yield = []
    monthly_bond_return = []
    squared_deviations = []
    
    df_cur = df.loc[df['STOCK CODE'] == stock, :]

    # create bond_returns list to store bond return of every month
    bond_returns = []
    
    for index in range(df_cur.shape[0]):

        yield_income.append(get_yield_income(df_cur.iloc[index, df_cur.columns.get_loc('NEXT COUPON RATE')], df_cur.iloc[index, df_cur.columns.get_loc('EVAL MID PRICE')]))
        roll_down_return.append(get_roll_down_return(df_cur.iloc[index, df_cur.columns.get_loc('EVAL MID PRICE CHANGE')], df_cur.iloc[index, df_cur.columns.get_loc('EVAL MID PRICE')]))
        expected_change_in_yield.append(get_expected_change_in_yield(df_cur.iloc[index, df_cur.columns.get_loc('MODIFIED DURATION')], df_cur.iloc[index, df_cur.columns.get_loc('EVAL MID YIELD CHANGE')], df_cur.iloc[index, df_cur.columns.get_loc('CONVEXITY')]))
        

    for i in range(len(yield_income)):

        monthly_bond_return.append(get_bond_return_of_month(yield_income[i], roll_down_return[i], expected_change_in_yield[i]))

    # monthly avg bond return for each stock
    mar = sum(monthly_bond_return) / df_cur.shape[0]
    monthly_avg_return.append(mar) 
          
    # calculate deviations for volatility
    for cur_return in monthly_bond_return:
        deviation = abs(mar - cur_return)
        squared_deviations.append(deviation * deviation)
                            
    # sum of all squared deviations                        
    sum_sd = sum(squared_deviations)
    v = math.sqrt(sum_sd / (df_cur.shape[0] - 1))   
    volatility.append(v)

    ratio = mar / v
    bond_over_volatility_ratio.append(ratio)
    
    # for checking purpose
#     if stock == 'DJ140003':
#         print(v)
#         print(f'{stock} | Bond Return: {mar} | Volatility: {v} | Ratio: {bond_over_volatility_ratio}')
    print(f'{stock} | Bond Return: {mar} | Volatility: {v} | Ratio: {ratio}')

# print(monthly_avg_return)
# print(volatility)
print(f'Total number of bonds: {len(volatility)} ')
```

```python
result_df = pd.DataFrame(
    {'Stock Code': stock_code,
     'Monthly Average Return': monthly_avg_return,
     'Volatility': volatility,
     'Return/ Volatility Ratio': bond_over_volatility_ratio
    })
result_df
```

```python
result_df = result_df.dropna()
sorted_result_df = result_df.sort_values('Return/ Volatility Ratio', ascending = False)
sorted_result_df
```
